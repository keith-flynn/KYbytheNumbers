```python
dollar = pd.read_csv('assets/RealTrade-WeightedValueoftheDollarforKY.csv')
dollar['DATE'] = pd.to_datetime(dollar['DATE'])

dollar.rename(columns={"RTWVDKY684NMFRBDAL":"VALUE"}, inplace=True)
print(dollar.head())

fig, ax = plt.subplots()
ax.axhline(100, linestyle='--', color='darkslategrey')
dollar_x = dollar['DATE']
dollar_y = dollar['VALUE']
plt.plot_date(dollar_x, dollar_y, fmt=',:', color='palegreen')
plt.gcf().autofmt_xdate()
plt.title('Real Trade-Weighted Value of the Dollar in Kentucky')
plt.ylabel('Value per Cent')
plt.show()
```
